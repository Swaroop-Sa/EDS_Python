```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/winequality-red.csv')

print(data.head())

print(data.shape)


#1. Calculate the mean of each column:

column_means = data.mean()
print("1.Mean of each column :",column_means)


#2. Calculate the standard deviation of each column:

column_std = data.std()
print("2.Standard deviation of each column :",column_std)


#3. Generate descriptive statistics of the dataset:

print("3.Descriptive statistics of the dataset:",data.describe())


#4. Count the number of missing values in each column:

missing_values = data.isnull().sum()
print("4.The number of missing values in each column:",missing_values)


#5. Replace missing values with the mean of each column:

data_filled = data.fillna(data.mean())
print("5.New dataset :",data_filled)


#6. Select a specific column from the DataFrame:

selected_column = data['quality']
print("6.Selected column :",selected_column)


#7. Filter rows based on a condition:

filtered_data = data[data['quality'] > 7]
print("7.Filtered rows based on a condition ['quality'] > 7 :",filtered_data)


#8. Sort the DataFrame by a specific column:

sorted_data = data.sort_values('alcohol')
print("8.Sorted data by column alcohol :",sorted_data)


#9. Group the data by a specific column and calculate the mean of each group:

grouped_data = data.groupby('quality').mean
print("9.Mean of column named 'quality' :",grouped_data)


#10. Create a histogram of a specific column using Matplotlib:

print("10.A histogram of a specific column using Matplotlib :")

plt.hist(data['alcohol'])
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.title('Histogram of Alcohol')
plt.show()


#11. Create a scatter plot of two columns using Matplotlib:

print("11.A scatter plot of two columns using Matplotlib :")

plt.scatter(data['density'], data['quality'])
plt.xlabel('density')
plt.ylabel('Quality')
plt.title('Scatter Plot of Density vs Quality')
plt.show()


#12. Calculate the correlation matrix of the dataset:

correlation_matrix = data.corr()
print("12.The correlation matrix of the dataset :",correlation_matrix)


#13. Create a heatmap of the correlation matrix using Matplotlib:

print("13.A heatmap of the correlation matrix using Matplotlib :")

plt.imshow(correlation_matrix, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.title('Correlation Matrix')
plt.show()


#14. Calculate the median of each column:

column_medians = data.median()
print("14.The median of each column :",column_medians)


#15. Calculate the maximum value in each column:

column_max = data.max()
print("15.The maximum value in each column :",column_max)


#16. Calculate the minimum value in each column:

column_min = data.min()
print("16.The minimum value in each column :",column_min)


#17. Calculate the sum of each column:

column_sum = data.sum()
print("17.The sum of each column :",column_sum)


#18. Select rows based on multiple conditions:

filtered_data = data[(data['alcohol'] > 12) & (data['quality'] > 7)]
print("18.Selected rows :",filtered_data)


#19. Calculate the average quality rating:

average_quality = data['quality'].mean()
print("19.The average quality rating :",average_quality)


#20. Count the number of unique values in a column:

unique_values = data['quality'].nunique()
print("20.The number of unique values in a column :",unique_values)


#21. Group the data by a specific column and calculate the maximum value in each group:

grouped_data = data.groupby('quality').max()
print("21.Grouped data :",grouped_data)

#22. Calculate the cumulative sum of a specific column:

cumulative_sum = data['sulphates'].cumsum()
print("22.The cumulative sum of a specific column :",cumulative_sum)


#23. Calculate the mode of a specific column:

column_mode = data['density'].mode()
print("23.The mode of a specific column :",column_mode)


#24. Calculate the skewness of a specific column:

column_skewness = data['volatile acidity'].skew()
print("24.The skewness of a specific column :",column_skewness)


#25. Calculate the kurtosis of a specific column:

column_kurtosis = data['residual sugar'].kurtosis()
print("25.The kurtosis of a specific column :",column_kurtosis)


#26. Replace values in a specific column based on a condition:

print("Newly created data :")
data.loc[data['density'] > 1.0, 'density'] = 1.0

```
