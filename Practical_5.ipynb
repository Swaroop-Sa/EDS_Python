```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 3: Read the CSV file
data = pd.read_csv('nutrients.csv')

print ("Graph 1: Bar graph for calorie distribution")
plt.figure(figsize=(10, 6))
plt.bar(data['Food'], data['Calories'], color='skyblue', edgecolor='black')
plt.xlabel('Food')
plt.ylabel('Calories')
plt.title('Calorie Distribution')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Graph 2: Pie chart for protein distribution
print ("Graph 2: Pie chart for protein distribution")
plt.figure(figsize=(8, 8))
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC', '#B2FF66', '#FF6666']
plt.pie(data['Protein'], labels=data['Food'], autopct='%1.1f%%', colors=colors)
plt.title('Protein Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()

# Graph 3: Line graph for fat content
print ("Graph 3: Line graph for fat content")
plt.figure(figsize=(10, 6))
plt.plot(data['Food'], data['Fat'], marker='o', color='green', linestyle='-', linewidth=2)
plt.xlabel('Food')
plt.ylabel('Fat (grams)')
plt.title('Fat Content')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Graph 4: Stacked bar graph for saturated fat and fiber
print ("Graph 4: Stacked bar graph for saturated fat and fiber")
plt.figure(figsize=(10, 6))
plt.bar(data['Food'], data['Sat.Fat'], label='Saturated Fat', color='red')
plt.bar(data['Food'], data['Fiber'], bottom=data['Sat.Fat'], label='Fiber', color='green')
plt.xlabel('Food')
plt.ylabel('Grams')
plt.title('Saturated Fat and Fiber')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# Graph 5: Scatter plot for carbs and calories
print ("Graph 5: Scatter plot for carbs and calories")
plt.figure(figsize=(8, 6))
plt.scatter(data['Carbs'], data['Calories'], color='purple', edgecolors='black', alpha=0.7)
plt.xlabel('Carbs (grams)')
plt.ylabel('Calories')
plt.title('Carbs vs Calories')
plt.tight_layout()
plt.show()


# Graph 6: Grouped bar graph for protein and fiber
print ("Graph 6: Grouped bar graph for protein and fiber")
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = data.index
plt.bar(index, data['Protein'], bar_width, label='Protein', color='orange')
plt.bar(index + bar_width, data['Fiber'], bar_width, label='Fiber', color='green')
plt.xlabel('Food')
plt.ylabel('Grams')
plt.title('Protein and Fiber Comparison')
plt.xticks(index + bar_width / 2, data['Food'], rotation=90)
plt.legend()
plt.tight_layout()
plt.show()


```
