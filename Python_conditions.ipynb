**Python Conditions , " if-elif-else " statements :**

**If :**

1.   An "if statement" is written by using the if keyword.

```python
#SYNTAX :
a = 33
b = 200
if b > a:
  print("b is greater than a")
```

**INDENTATION : (Whitespace at the beginning of a line)**

1.   If statement, without indentation (will raise an error):


```python
a = 33
b = 200
if b > a:
print("b is greater than a") # you will get an error

#SAME RULE applies in the syntax of ' Elif ' and 'Else' statements !
```

**Elif :**

1.  The elif keyword is Python's way of saying "if the previous conditions were not true, then try this condition".


```python
#SYNTAX :

a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

**Else :**

1.  The else keyword catches anything which isn't caught by the preceding conditions.

```python
#SYNTAX :

a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:                    # You can also have an else without the elif
  print("a and b are equal")
else:
  print("a is greater than b")
```

**Nested if-else :**

1.  You can have ' if statements inside if statements ', this is called nested if statements.

```python
#SYNTAX:

x = 15

if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
  else:
    print("but not above 20.")
```

**Short Hand If :**

1.   If you have only " one statement " to execute, you can put it on the same line as the if statement

```python
#SYNTAX :

a = 200
b = 33

if a > b: print("a is greater than b")
```

**Short Hand If ... Else :**

1.   If you have only " one statement " to execute, one for if, and one for else, you can put it all on the same line:

```python
#SYNTAX :

a = 2
b = 330
print("A") if a > b else print("B")
```

```python
x = 10
y = "Even" if x % 2 == 0 and x > 5 else "Odd"   #More complex and less readable.
print(y)


#Use traditional ' if-else ' statement syntax.

x = 10
if x % 2 == 0 and x > 5:
    y = "Even"
else:
    y = "Odd"
print(y)
```

**Logical operators and Conditional statements :**

**And**
*   The ' and ' keyword is a logical operator, and is used to combine conditional statements:

```python
#SYNTAX :

a = 200
b = 3300
c = 500
if a > b and c > a:
  print("Both conditions are True")
else:
  print(False)
```

**Or**
*   The ' or ' keyword is a logical operator, and is also used to combine conditional statements:

```python
#SYNTAX :

a = 200
b = 3300
c = 500
if a > b or a > c:
  print("At least one of the conditions is True")
else:
  print(False)
```

**Not**

*  The ' not ' keyword is a logical operator, and is used to reverse the result of the conditional statement:

```python
#SYNTAX :

a = 33
b = 200
if not a > b:
  print("a is NOT greater than b")
else:
  print("a is greater than b")  #Example
```
