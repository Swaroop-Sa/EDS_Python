```python

#mse practical


import csv
file=open("stud1.csv","r")
file1=list(csv.reader(file, delimiter = ','))

gen=[]
race=[]
mscore=[]
tpc=[]

while True:
  data=file.readline()
  if not data:
    break

gen.append(file1[0])
race.append(file1[1])
mscore.append(file1[4])
tpc.append(file1[3])

file.close()

print("Gender : ",gen)
print("Race : ",race)
print("Maths score : ",mscore)
print("test preparation course : ",tpc)

#*****QUERIES*****#

#1.How many female candidates are in group A:
c=0
for i in range(1,len(file1)):
  if (file1[i][1]=="group A") and (file1[i][0]=="female"):
    c+=1
print(c)
#2.How many male candidates are in group B:
d=0
for i in range(1,len(file1)):
  if (file1[i][1]=="group B") and (file1[i][0]=="male"):
    d+=1
print(d)

#3.How many candidates are in group C:
e=0
for i in range(1,len(file1)):
  if (file1[i][1]=="group C") :
    e+=1
print(e)

#4.How many students are having score greater than 80 in mathematics:
f=0
for i in range(1,len(file1)):
  if (int(file1[i][4]) > 80) :
    f+=1
print(f)

#5.How many students have completed the test preparation course:
g=0
for i in range(1,len(file1)):
  if (file1[i][3]=="completed") :
    g+=1
print(g)
```
